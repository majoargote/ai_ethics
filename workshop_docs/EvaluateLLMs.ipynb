```python
import requests
from IPython.display import Markdown, display

url = "http://127.0.0.1:11434/api/chat"
payload = {
    "model": MODEL_NAME,
    "messages": [
        {"role": "user", "content": "Write a short hello message."}
    ],
    "stream": False,
}

resp = requests.post(url, json=payload, timeout=300)
resp.raise_for_status()

data = resp.json()

display(Markdown(f"### Content\n{data['message']['content']}"))

if "thinking" in data["message"]:
    display(Markdown(f"### Thinking\n{data['message']['thinking']}"))
```
